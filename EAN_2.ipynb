#### *Limpieza con if en el DF*

Dado que tenemos varias columnas con simbolos y guiones en el Df, vamos a realizar una iteracion para: remover "-", remover "$"

```python
# Seleccionar columnas para reemplazar
columnas_reemplazo = ['REGIÓN', 'DEPARTAMENTO DOMICILIO','CIUDAD DOMICILIO','INGRESOS OPERACIONALES','GANANCIA (PÉRDIDA)',
                      'TOTAL ACTIVOS','TOTAL PASIVOS','TOTAL PATRIMONIO']

# Iteracion y remplazo de simbolo "-"
for col in columnas_reemplazo:
    df_empresas[col] = df_empresas[col].str.replace("-", " ").str.replace("$", "")

# Reemplazo de puntos por comas opcional
#columnas_valores = ['INGRESOS OPERACIONALES','GANANCIA (PÉRDIDA)','TOTAL ACTIVOS','TOTAL PASIVOS','TOTAL PATRIMONIO']
#for col in columnas_valores:
#    df_empresas[col] = df_empresas[col].str.replace(".", ",")

df_empresas
```

## Actividad 2. Transformación de datos.

### Actividades especificas guia 2:

#### Reducción de los nits de longitud mayor a 9.

```python
# TRANSFORMACION DE DATOS

# Convertir la columna 'NIT' a cadenas de texto temporalmente para medir la longitud
df_empresas['NIT_str'] = df_empresas['NIT'].astype(str)

# Filtrar valores con longitud mayor a 9
df_empresas_reducido = df_empresas[df_empresas['NIT_str'].apply(len) <= 9]

# Eliminar columna temporal 'NIT_str'
df_empresas_reducido.drop(columns=['NIT_str'], inplace=True)

df_empresas_reducido

```

### Convertir tipo variable

```python
# Lista de columnas que necesitamos convertir a float
columnas_a_convertir = ['INGRESOS OPERACIONALES', 'GANANCIA (PÉRDIDA)', 'TOTAL ACTIVOS', 'TOTAL PASIVOS', 'TOTAL PATRIMONIO']

# Convertir columnas predefinidas a float
df_empresas[columnas_a_convertir] = df_empresas[columnas_a_convertir].replace(",", "", regex=True)  # Eliminar comas si existen
df_empresas[columnas_a_convertir] = df_empresas[columnas_a_convertir].astype(float)

# Verificar que las columnas han sido convertidas a float
print(f"tipo de variable columnas convertidas :\n\n{df_empresas[columnas_a_convertir].dtypes}")
print(f"\nTipos de variable del dataset actual:\n\n{df_empresas.dtypes}")
```

* TOTAL ACTIVOS y TOTAL PATRIMONIO (0.96): Una alta correlación entre activos y patrimonio indica que, a medida que aumentan los activos de una empresa, también suele aumentar su patrimonio.
Correlaciones Moderadas a Fuertes:

* TOTAL ACTIVOS y INGRESOS OPERACIONALES (0.91): Existe una fuerte correlación positiva entre los activos totales y los ingresos operacionales, lo que sugiere que las empresas con más activos tienden a generar más ingresos.
* TOTAL PATRIMONIO y GANANCIA (PÉRDIDA) (0.85): También hay una fuerte relación entre el patrimonio total y las ganancias, indicando que las empresas con mayor patrimonio suelen ser más rentables.
Correlación Moderada:

* INGRESOS OPERACIONALES y TOTAL PATRIMONIO (0.81): Aunque la correlación no es tan fuerte como en los casos anteriores, sigue siendo positiva y significativa. ### Calculo de crecimiento por sectores

```python
import pandas as pd
import matplotlib.pyplot as plt

df_empresas['Año de Corte'] = pd.to_numeric(df_empresas['Año de Corte'], errors='coerce')

# Agrupación por "MACROSECTOR" y "Año de Corte" sumando los "INGRESOS OPERACIONALES"
df_sector_anual = df_empresas.groupby(['MACROSECTOR', 'Año de Corte'])['INGRESOS OPERACIONALES'].sum().reset_index()

df_sector_anual.head()
```

```python
import seaborn as sns

# Crear el gráfico de líneas
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_sector_anual, x='Año de Corte', y='INGRESOS OPERACIONALES', hue='MACROSECTOR', marker='o')

# Configuración del gráfico
plt.title('Crecimiento de los Ingresos Operacionales por Sector el ultimo Año')
plt.xlabel('Año de Corte')
plt.ylabel('Ingresos Operacionales (millones)')
plt.legend(title='Macrosector')
plt.grid(True)

plt.show()
```
