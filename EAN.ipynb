#### *Limpieza con if en el DF Titanic*

Dado que no tenemos nulos en el Df, vamos a realizar una iteracion para asignar una categoria de tarifa a los pasajeros basado en su Edad

```python
# Iteracion en las filas del dataframe
for index, row in df_titanic.iterrows():
    # rellenar nulos con la edad promedio
    if pd.isnull(row['Age']):
        df_titanic.at[index, 'Age'] = df_titanic['Age'].mean()

    # Correccion de valores menores a cero con ceros
    if row['Fare'] < 0:
        df_titanic.at[index, 'Fare'] = 0
    # Categorizacion basada en edad    
    if row['Age'] < 18:
        df_titanic.at[index, 'Categoria'] = 'Ni√±o'
    elif row['Age'] < 60:
        df_titanic.at[index, 'Categoria'] = 'Adulto'
    else:
        df_titanic.at[index, 'Categoria'] = 'Tercera Edad'
        
df_titanic
```

#### *Limpieza con if en el DF Income*

Dado que tenemos varias columnas con nulos en el Df, vamos a realizar una limpieza y adicionalmente una iteracion para: remover "-", remover "." y asignar los ingresos en la columna income a una categoria de ingesos "low" y "high"

```python
df_income["income"].unique()
```

```python
# Remover columnas con valores nulos
df_income.dropna(inplace=True)

# Seleccionar columnas para reemplazar
columnas_reemplazo = ['workclass','education','marital-status','occupation','relationship','race','native-country']

# Iteracion y remplazo de simbolo "-"
for col in columnas_reemplazo:
    df_income[col] = df_income[col].str.replace("-", " ")

# Limpieza de columna income opcional
df_income['income'] = df_income['income'].str.replace(".", "", regex=False)

# Opcion 1 de categoria con mapeo
#df_income['income'] = df_income['income'].map({'<=50K': 'low', '>50K': 'high'})

# Opcion 2 de categoria con if
for index, row in df_income.iterrows():
    if row['income'] == '<=50K' or row['income'] == '<=50K.':
        df_income.at[index, 'income'] = 'low'
    elif row['income'] == '>50K' or row['income'] == '>50K.':
        df_income.at[index, 'income'] = 'high'

df_income
```
