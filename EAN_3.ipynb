#### *Limpieza con if en el DF*

Dado que tenemos varias columnas con simbolos y caracteres en el Df, vamos a realizar una iteracion para: remover "-", remover "$" y texto no deseado

### Limpieza iterativa por columnas

```python
# Seleccionar columnas para reemplazar
columnas_reemplazo = ['valor_presupuesto_general','presupuesto_menor_cuantia', 'presupuesto_minima_cuantia']

# Iteracion y remplazo de simbolo "-"
for col in columnas_reemplazo:
    df_concatenado[col] = df_concatenado[col].str.replace("-", "").str.replace("$", "")

# Reemplazo de puntos por comas opcional
#columnas_valores = ['valor_presupuesto_general','presupuesto_menor_cuantia', 'presupuesto_minima_cuantia']
#for col in columnas_valores:
#    df_concatenado[col] = df_concatenado[col].str.replace(".", ",")

df_concatenado
```

### Transformacion de formato iterativo por columnas

Identificación y alguna transformación del tipo de dato.

```python
columnas_a_verificar = ['valor_presupuesto_general','presupuesto_menor_cuantia', 'presupuesto_minima_cuantia']

#Desactivar la notacion cientifica
pd.set_option('display.float_format', '{:.2f}'.format)

# Convertir las columnas a numéricas, ignorando los errores
for col in columnas_a_verificar:
    df_concatenado[col] = pd.to_numeric(df_concatenado[col], errors='coerce')

# Filtro para mantener solo las filas con valores numéricos
filtro_numerico = df_concatenado[columnas_a_verificar].notnull().all(axis=1)

# Filtro al DataFrame original para eliminar las filas con valores no numéricos
df_final = df_concatenado[filtro_numerico]

df_final
```

### Utilizar al menos 2 tipos de gráficos para variables seleccionadas. 

##### Grafica stacked para mostrar los departamentos con mas presupuesto a lo largo de los años, con el valor de presupuesto sumado por departamento

```python
# Eliminar filas con valor_presupuesto_general igual a 0
df_sin_2025 = df_sin_2025[df_sin_2025['valor_presupuesto_general'] > 0.5]


# Agrupar los datos por Año y departamento_paa, sumando el valor_presupuesto_general
df_agrupado = df_sin_2025.groupby(['Año', 'departamento_paa'])['valor_presupuesto_general'].sum().reset_index()
df_agrupado

```

```python
conteo = df_agrupado['valor_presupuesto_general'].value_counts()
conteo
```

```python
import plotly.express as px

# Crear la gráfica de barras apiladas
fig = px.bar(df_agrupado, 
             x='Año',
             y='valor_presupuesto_general',
             color='departamento_paa',
             title='Presupuesto General por Año y Departamento',
             labels={'valor_presupuesto_general': 'Presupuesto General', 'departamento_paa': 'Departamento'},
             barmode='stack',
             text_auto=True  # Mostrar valores en las barras
            )

fig.show()
```

##### Grafica de dona para visualizar top 5 empresas con mas presupuesto a nivel general

```python
# Seleccionar las 5 empresas con el mayor presupuesto general
top_5_empresas = df_final[['nombre_entidad', 'valor_presupuesto_general', 'Año']].sort_values(by='valor_presupuesto_general', ascending=False).head(5)

top_5_empresas
```

```python
# Crear el gráfico de donut
fig = px.pie(top_5_empresas, 
             names='nombre_entidad', 
             values='valor_presupuesto_general', 
             title='Top 5 Empresas con Mayor Presupuesto General',
             hole=0.4)  # Agujero en el centro

fig.show()
```

##### Finalmente un grafico de lineas en matplotlib para visualizar los presupuestos en total a travez de los años

```python
# Agrupar los datos por año y sumar el presupuesto general
df_presupuesto_por_ano = df_sin_2025.groupby('Año')['valor_presupuesto_general'].sum().reset_index()

print(df_presupuesto_por_ano)
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(df_presupuesto_por_ano['Año'], df_presupuesto_por_ano['valor_presupuesto_general'], marker='o')

plt.title('Presupuesto Total por Año', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Presupuesto General (en millones)', fontsize=12)

plt.xticks(df_presupuesto_por_ano['Año'], rotation=45)

plt.grid(True)
plt.show()
```

```python
# Filtrar por las 10 entidades con mayor presupuesto
top_10_entidades = df_final.groupby('nombre_entidad')['valor_presupuesto_general'].sum().nlargest(10).reset_index()

# Crear gráfico de barras
fig = px.bar(top_10_entidades, 
             x='nombre_entidad', 
             y='valor_presupuesto_general', 
             title='Top 10 Entidades con Mayor Presupuesto General')

fig.show()
```
